```python
#import the file, the code is the same as the other notebook but i used all potential wordle answers instead of all allowed guesses
with open("wordle-answers-alphabetical.txt") as f:
    file = f.read()
```

```python
#output
file
```

```python
#as you can see, each word is separated by the "\n" character, so we will split the string on that character
splitted_word_list = file.split("\n")
#perfect!!
splitted_word_list
```

```python
import string
alphabet = string.ascii_lowercase
```

```python
alphabet
```

```python
#get all the unique characters of the alphabet and store them all in a dictionary/set
alphabet_dict = set(alphabet)
print(alphabet_dict)
```

```python
import numpy as np
#create an empty list which will store our final results and populate it with 0s
final_results = np.zeros(len(alphabet_dict))
```

```python
for index,character in enumerate(alphabet_dict):
    print(character)
    for word in splitted_word_list:
        #loop through each word in the word list
        for letter in word:
            #loop through each letter in each word
            if letter == character:
                final_results[index] += 1
```

```python
#right now our final_results list has float(decimal) values, so we convert all elements to integers
integer_letter_counts = final_results.astype("int64")
```

```python
print(integer_letter_counts)
#create a new empty list which we will use to store our findings
tup_list = []
```

```python
#combine the letter counts with the set containing the alphabet
for i,letter in enumerate(alphabet_dict):
    tup_list.append((letter,integer_letter_counts[i]))
```

```python
#create a list containing elements in the form: (letter,frequency) - we preapre it this way to use the dict() function on it
#which will convert this list into a dictionary
tup_list
```

```python
#convert this list into a dictionary using dict()
dict(tup_list)
```

```python
#sort the final dictionary based on frequency, in descending order
dict(sorted(dict(tup_list).items(), key=lambda item: item[1],reverse=True))
```
